```python
# PART 1 - TASK1
print("="*50)
print("Part 1 - Task1 — Data Parsing & Profile Cleaning")
print("="*50)

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma", "roll": "102", "marks_str": "55, 68, 49, 72, 60"},
    {"name": "  Priya Nair ", "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA", "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ", "roll": "105", "marks_str": "75, 80, 70, 68, 82"}
]
# Task1.1
cleaned_students = []

for student in raw_students:
    name = student["name"].strip().title()
    roll = int(student["roll"])
    marks = list(map(int, student["marks_str"].split(",")))

# Task1.2

    if all(word.isalpha() for word in name.split()):
        validity = " ✅ Valid name"
    else:
        validity = " ❌ Invalid name"

    cleaned_students.append({
        "name": name,
        "roll": roll,
        "marks": marks
    })

# Task1.3

    print(f"Student : {name} ({validity})")
    print(f"Roll No : {roll}")
    print(f"Marks   : {marks}")
    print("="*50)

# Task1.4

for student in cleaned_students:
    if student["roll"] == 103:
        print(student["name"].upper())
        print(student["name"].lower())


################################################################################



# PART 1 - TASK2

print("="*50)
print("Part 1 - Task2 - Marks Analysis Using Loops & Conditionals")
print("="*50)

# Task2.1

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

for sub,mark in zip(subjects,marks):
  if mark >= 90:
    print(sub,",", mark,",","Grade:A+")
  elif 90> mark >=80:
    print(sub,",",mark,",","Grade:A")
  elif 80> mark >=70 :
    print(sub,",",mark,",","Grade:B")
  elif 70> mark >=60:
    print(sub,",",mark,",","Grade:C")
  elif 60> mark:
    print(sub,",",mark,",","Grade:F")

print("="*50)

# Task2.2
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

#Total marks
total_marks = sum(marks)
print("Total Marks:",total_marks)

#Average marks
avg_marks = total_marks/ len(marks)
print("Average Marks:",round(avg_marks, 2))

#Highest score
highest_marks = max(marks)
for sub,mark in zip(subjects,marks):
  if highest_marks == mark:
    print("Subject:",sub,", Highest Score:",highest_marks)

#Lowest score
lowest_marks = min(marks)
for sub,mark in zip(subjects,marks):
  if lowest_marks == mark:
    print("Subject:",sub,", Lowest Score:",lowest_marks)

# Task2.3

print("="*50)
new_count = 0
while True:
  subject = input("Enter the subject name(or done to stop): ")
  if subject == "done":
    break

  marks_input = input("enter the marks scored(0-100): ")
  if not marks_input.isdigit():
        print("do not crash, and do not add invalid entries to the list.")
        continue

  mark = int(marks_input)
  if mark < 0 or mark > 100:
        print("Marks must be between 0 and 100.")
        continue

  subjects.append(subject)
  marks.append(mark)
  new_count += 1

print("="*50)
print("new subject added: ",new_count)

new_total = sum(marks)
new_avg = new_total/len(marks)
print("updated average: ",round(new_avg,2))




################################################################################



# Part 1 - Task3

print("="*50)
print("Part 1 - Task 3 - Class Performance Summary")
print("="*50)

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

pass_list = []
fail_list = []
averages = []

# Task 3.1

for name, scores in class_data:
    avg = sum(scores) / len(scores)
    avg = round(avg, 2)

    if avg >= 60:
        status = "Pass"
        pass_list.append(name)
    else:
        status = "Fail"
        fail_list.append(name)

    averages.append(avg)

    print(f"{name}: Average = {avg}, Status = {status}")

# Task 3.2

print("="*50)
print(f"{'Name':<15} | {'Average':<7} | Status")
print("="*50)

for i in range(len(class_data)):
    name = class_data[i][0]
    avg = averages[i]
    if avg >= 60:
      status = "Pass"
    else:
      status= "Fail"

    print(f"{name:<15} | {avg:<7} | {status}")

# Task 3.3

print("="*50)

print("Students Passed:", pass_list)
print("Students Failed:", fail_list)

class_avg = sum(averages) / len(averages)
print("Class Average:", round(class_avg, 2))




################################################################################




print("="*50)
print("Part 1 - Task4 - String Manipulation Utility")
print("="*50)

essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "


# Task 4.1
clean_essay = essay.strip().lower()
print("Cleaned Essay:", clean_essay)

# Task 4.2

print("="*50)
print("Title Case:")
print(clean_essay.title())

# Task 4.3

print("="*50)
count_python = clean_essay.count("python")
print("Count of 'python':", count_python)

# Task 4.4

print("="*50)
replaced_essay = clean_essay.replace("python", "Python 🐍")
print("Replaced Essay:")
print(replaced_essay)

# Task 4.5

print("="*50)
sentences = clean_essay.split(". ")
print("Sentence List:")
print(sentences)

# Task 4.6

print("="*50)
print("Numbered Sentences:")
for i in range(len(sentences)):
    sentence = sentences[i]
    if not sentence.endswith("."):
        sentence += "."

    print(f"{i+1}. {sentence}")
```
